```python
import numpy as np
import pandas as pd
```

```python
Interests = pd.read_csv('Topic_Survey_Assignment.csv',index_col = 0)
Interests
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
%matplotlib inline
```

```python
Interests.sort_values(by='Very interested', ascending=False, inplace=True)
```

```python
total_resp = 2233
```

```python
for col in Interests.columns:
    Interests[col] /= total_resp
Interests
```

```python
ax = Interests.plot(kind='bar',figsize=(20,8),width=0.8,color=['#5cb85c','#5bc0de','#d9534f'])
plt.title("Percentage of Respondents' Interest in Data Science Areas",fontsize= 16)

legend = plt.legend(labels=Interests.columns,fontsize= 14,frameon = 1)
frame = legend.get_frame()
frame.set_facecolor('white')
frame.set_edgecolor('black')

ax.set_facecolor('xkcd:white')

plt.xticks(fontsize=14)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.yticks([])

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:.0%}'.format(height), (x, y + height + 0.01))
```

```python
crime = pd.read_csv('https://cocl.us/sanfran_crime_dataset')
crime
```

```python
total = pd.DataFrame(crime['PdDistrict'].value_counts())
total
```

```python
final = pd.DataFrame({'Neighborhood':total.index.values,'Count': total.PdDistrict})
final.index = range(0,10)
final
```

```python
import folium
```

```python
san_geo = r'san-francisco.geojson'
```

```python
san_map = folium.Map(location = [37.774,-122.419],zoom_start = 12)
san_map.choropleth(
    geo_data=san_geo,
    data=final,
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco'
)
san_map
```
